```python
from pathlib import Path
import pandas as pd
import re
import math

import matplotlib.pyplot as plt

final_df = pd.DataFrame()
# TODO: Add back "Num Bedrooms" if we have time 
final_df = final_df.reindex(columns=["Year", "Longitude", "Latitude", "Income", "Age", "Population", "Num Rooms", "House Value"])

zipcode_df = pd.read_csv("./Bitcamp_Housing_Data_Files/Output_House_Value/ACSDT5Y2011.B25077-Data.csv")
zipcodes = []
for zipcode in zipcode_df['GEO_ID'].values:
    zipcodes.append(zipcode[9:])
del zipcodes[0]
print(zipcodes)

# zipcode_df = pd.read_csv("./Bitcamp_Housing_Data_Files/Output_House_Value/ACSDT5Y2021.B25077-Data.csv")
# zipcodes = []
# for zipcode in zipcode_df['GEO_ID'].values:
#     zipcodes.append(zipcode[9:])
# del zipcodes[0]
# print(zipcodes)

# value_df = pd.read_csv(f"./Bitcamp_Housing_Data_Files/Output_House_Value/ACSDT5Y2011.B25077-Data.csv")
# value = value_df.loc[value_df['GEO_ID'] == f"8600000US89010"]["B25077_001E"].iloc[0]
# print(value)

for year in range(2011, 2021):
    value_df = pd.read_csv(f"./Bitcamp_Housing_Data_Files/Output_House_Value/ACSDT5Y{year}.B25077-Data.csv")
    income_df = pd.read_csv(f"./Bitcamp_Housing_Data_Files/Input1_Income/ACSST5Y{year}.S1901-Data.csv")
    age_df = pd.read_csv(f"./Bitcamp_Housing_Data_Files/Input2_Age/ACSST5Y{year}.S0101-Data.csv")
    population_df = pd.read_csv(f"./Bitcamp_Housing_Data_Files/Input2_Age/ACSST5Y{year}.S0101-Data.csv")
    rooms_df = pd.read_csv(f"./Bitcamp_Housing_Data_Files/Input4_Median_Rooms_Median_Bedrooms/ACSDP5Y{year}.DP04-Data.csv")
    latitude_longitude_df = pd.read_csv(f"./Bitcamp_Housing_Data_Files/Input5_Latitutude_Longitude/zip_code_database.csv")

    for zipcode in zipcodes:
        value = value_df.loc[value_df['GEO_ID'] == f"8600000US{zipcode}"]["B25077_001E"].iloc[0]
        income = income_df.loc[income_df['GEO_ID'] == f"8600000US{zipcode}"]["S1901_C01_012E"].iloc[0]
        age = age_df.loc[age_df['GEO_ID'] == f"8600000US{zipcode}"]["S0101_C01_030E"].iloc[0]
        population = population_df.loc[population_df['GEO_ID'] == f"8600000US{zipcode}"]["S0101_C01_001E"].iloc[0]
        rooms = rooms_df.loc[rooms_df['GEO_ID'] == f"8600000US{zipcode}"]["DP04_0036E"].iloc[0]
        latitude = latitude_longitude_df.loc[latitude_longitude_df['zip'] == int(zipcode)]["latitude"].iloc[0]
        longitude = latitude_longitude_df.loc[latitude_longitude_df['zip'] == int(zipcode)]["longitude"].iloc[0]

        if any(char in income for char in ",+"):
            income = re.sub(r'\W+', '', income)


        if not (type(value) is float and math.isnan(value)) and any(char in value for char in ",+"):
            value = re.sub(r'\W+', '', value)
        
        final_df.loc[len(final_df.index)] = [year, longitude, latitude, income, age, population, rooms, value]

final_df.head()
```
